สร้างตารางที่แสดงรายการเลือกตั้งแบบแบ่งเขตเลือกตั้งและพรรคพวก เรียงตารางตามจำนวนโหวตที่ได้รับและแสดงเฉพาะ 10 อันดับแรก (30 คะแนน)

## 1.1 ตารางแสดงแบบแบ่งเขต

ทำการหาค่าผลรวมโหวตของแต่ละพรรคว่ามีผลโหวตรวมเป็นยังไง

```python
candidate = table_candidate.groupby('name').sum().reset_index()[['name','mp_app_vote']]
candidate
```

จากนั้นจะใช้จำนวนโหวตไปเรียงลำดับโดยใช้ list ของผลโหวตทั้งหมดไปหา

```python
len(list(candidate['mp_app_vote']))
```

loop ด้วยการ sort ด้วยวิธี bubble sort

```python
list_can1 = list(candidate['mp_app_vote'])
n = len(list_can1)

for i in range(n - 1):
    for j in range(0, n - i - 1):
        if list_can1[j] < list_can1[j + 1]:
            list_can1[j], list_can1[j + 1] = list_can1[j + 1], list_can1[j]

print(list_can1)
```

จากนั้นนำ list ที่ได้ไปสร้างตารางเป็น dataframe อันใหม่

```python
dfcan1 = pd.DataFrame(list_can1, columns=["vote"])
dfcan1
```

แล้วเอา dataframe ไปเทียบกับตารางพรรคนั้นๆว่าอยู่พรรคไหนจำนวน 10 พรรคแรก

```python
dfcan1['name'] = dfcan1['vote'].map(candidate.set_index('mp_app_vote')['name'])
dfcan1.head(10)
```

สรุปว่าในการเลือกตั้งแบบแบ่งเขต พบว่า พรรคที่ได้คะแนนมากที่สุด 10 อันดับแรกเรียงตามอันดับได้แก่

2. เพื่อไทย
3. ภูมิใจไทย
4. พลังประชารัฐ
5. รวมไทยสร้างชาติ
6. ประชาธิปัตย์
7. ไทยสร้างไทย
8. ชาติไทยพัฒนา
9. ประชาชาติ
10. ชาติพัฒนากล้า

```python
dfpar1 = pd.DataFrame(list_par1, columns=["vote"])
dfpar1
```
